```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab import drive
drive.mount('/content/drive')
```

```python
# ZAD 1 Wydziel obraz kanalu G

path = testImagePath = '/content/drive/MyDrive/rozpoznawanie/maxresdefault.jpg'
img = Image.open(path)
img = img.convert('RGB')

green_shades = img.load()
for i in range(img.size[0]):
    for j in range(img.size[1]):
        r, g, b = green_shades[i, j]
        grayscale = int((r + g + b) / 3)
        green_shades[i, j] = (0, grayscale, 0)
img.save('/content/drive/MyDrive/rozpoznawanie/zielen.jpg')
```

```python
#ZAD 2
path = testImagePath = '/content/drive/MyDrive/rozpoznawanie/maxresdefault.jpg'
img = Image.open(path)
npImageArray = np.asarray(img , dtype='uint8')
R = 0.1
G = 0.2
B = 0.1
gray = R*(npImageArray[:,:,0])/3+G*(npImageArray[:,:,1])/3+B*(npImageArray[:,:,2])/3
imgGray = Image.fromarray(np.uint8(gray))
plotGray = plt.imshow(imgGray)
imgGraySave = imgGray.save('/content/drive/MyDrive/rozpoznawanie/gray.jpg')
```

```python
#ZAD 3
path = testImagePath = '/content/drive/MyDrive/rozpoznawanie/maxresdefault.jpg'
img = Image.open(path)
npImageArray = np.asarray(img , dtype='uint8')

R=(npImageArray[:,:,0])
G=(npImageArray[:,:,1])
B=(npImageArray[:,:,2])
imgGray = Image.fromarray(np.uint8(G+B+R))
plotGray = plt.imshow(imgGray)
imgGraySave = imgGray.save('/content/drive/MyDrive/rozpoznawanie/zad3.jpg')
```

```python
#ZAD 4
testImagePath = '/content/drive/MyDrive/rozpoznawanie/maxresdefault.jpg'
img = Image.open(testImagePath)
npImageArray = np.asarray(img , dtype='uint8')

R = npImageArray[:,:,0]
G = npImageArray[:,:,1]
B = npImageArray[:,:,2]
imgGrayArray = 0.2989 * R + 0.5870 * G + 0.1140 * B
imgGray = Image.fromarray(np.uint8(imgGrayArray))

plt.imshow(imgGray, cmap='gray')
plt.axis('off')
plt.show()

imgGray.save('/content/drive/MyDrive/rozpoznawanie/zad4.jpg')

oMin = np.min(npImageArray)
oMax = np.max(npImageArray)
mode_val = np.bincount(imgGrayArray.flatten().astype(int)).argmax()
plt.hist(imgGrayArray.flatten(), bins=256, range=[0, 255])
plt.xlabel('wartość')
plt.ylabel('liczebność')
plt.title('Histogram obrazu w odcieniach szarości')
plt.text(0.95, 0.75, f'Modalna: {mode_val}', transform=plt.gca().transAxes, verticalalignment='top', horizontalalignment='right')
plt.text(0.8, 0.6, f'Max: {oMax}', transform=plt.gca().transAxes, verticalalignment='top', horizontalalignment='right')
plt.text(0.6, 0.5, f'Min: {oMin}', transform=plt.gca().transAxes, verticalalignment='top', horizontalalignment='right')
plt.show()
```
